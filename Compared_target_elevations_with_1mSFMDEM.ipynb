```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
dirn="c:/jhigh/Projects/"
fn="Targets_elevations_comparison_with_1mSFMDEM.csv"
print(dirn + fn)
fullpathn = dirn + fn
```

```python
df=pd.read_csv(fullpathn, header = 0, skiprows=0, delimiter=',')
print("Read: ",len(df.index),' lines.')
df
```

```python
df['Elevation Difference'] = df['ELEV'] - df['Local_Z'] 
df 
```

```python
df.describe()
val = df["Elevation Difference"].values
rmse=np.sqrt(np.mean(val**2))
print(df.describe())
print("rmse=",rmse)
```

```python
df.hist(column= "Elevation Difference")
plt.savefig("Targets_vs_1mSFMDEM_Elevation_Difference_Histogram.png")
```

```python
df.plot(x='Local_Z', y='ELEV', style="o")
```

```python
df.plot(kind='scatter', x='Local_E', y='Local_N', c="Elevation Difference", cmap= "jet")
plt.savefig("Position_ScatterPlot_Targets_1mSFMDEM_ElevDif.png")
```

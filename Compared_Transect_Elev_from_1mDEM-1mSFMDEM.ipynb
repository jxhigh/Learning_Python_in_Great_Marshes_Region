```python
dirn="c:/jhigh/Projects/"
fn="Transect_points_elevations_from_1mDEM-1mSFMDEM_pointsdeleted.csv"
print(dirn + fn)
fullpathn = dirn + fn
```

```python
df=pd.read_csv(fullpathn, header = 0, skiprows=0, delimiter=',')
print("Read: ",len(df.index),' lines.')
df
```

```python
df.describe()
```

```python
binedges= np.arange(-.4,.45, .05)
binedges
```
